````python
# 5 –≤–∞—Ä—ñ–∞–Ω—Ç 

def is_prime(n):
    """–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞, —á–∏ —î —á–∏—Å–ª–æ –ø—Ä–æ—Å—Ç–∏–º"""
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

def kth_prime(k):
    """–ü–æ–≤–µ—Ä—Ç–∞—î k-—Ç–µ –ø—Ä–æ—Å—Ç–µ —á–∏—Å–ª–æ"""
    if not isinstance(k, int) or k < 1:
        raise ValueError("k –º–∞—î –±—É—Ç–∏ —Ü—ñ–ª–∏–º —á–∏—Å–ª–æ–º –±—ñ–ª—å—à–µ 0")

    count = 0
    num = 1
    while count < k:
        num += 1
        if is_prime(num):
            count += 1
    return num


try:
    k = int(input("–í–≤–µ–¥—ñ—Ç—å –Ω–æ–º–µ—Ä –ø—Ä–æ—Å—Ç–æ–≥–æ —á–∏—Å–ª–∞ (k): "))
    result = kth_prime(k)
    print(f"{k}-—Ç–µ –ø—Ä–æ—Å—Ç–µ —á–∏—Å–ª–æ: {result}")
except ValueError as e:
    print("–ü–æ–º–∏–ª–∫–∞:", e)

# –¢–µ—Å—Ç 1
print("5-—Ç–µ –ø—Ä–æ—Å—Ç–µ —á–∏—Å–ª–æ:", kth_prime(5)) 

# –¢–µ—Å—Ç 2
print("10-—Ç–µ –ø—Ä–æ—Å—Ç–µ —á–∏—Å–ª–æ:", kth_prime(10)) 

# –¢–µ—Å—Ç 3
print("20-—Ç–µ –ø—Ä–æ—Å—Ç–µ —á–∏—Å–ª–æ:", kth_prime(20))  
````
